```python
import pandas as  pd

%matplotlib inline
```

```python
df=pd.read_csv('../data/data.csv')

df = df.drop('ID', axis=1)
print(df.isnull().sum())
```

```python
# df.fillna(df.mean(), inplace=True)
df.fillna(value='5', inplace=True)
# df=df.fillna(method='pad')
```

```python
from sklearn.model_selection import train_test_split
X = df.drop('Response', axis=1)
y = df['Response']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20)
```

```python
from sklearn.ensemble import RandomForestRegressor,RandomForestClassifier
# Instantiate model with 100 decision trees
rf = RandomForestClassifier(n_estimators = 5,n_jobs=-1, random_state=42)
# Train the model on training data
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)
```

```python
import numpy as np
from math import sqrt
from sklearn.metrics import classification_report, accuracy_score,mean_squared_error, recall_score
errors = abs(y_pred - y_test)
mape = 100 * (errors / y_test)
mae= round(np.mean(errors),2)
accuracy = 100 - np.mean(mape)
rmse = sqrt(mean_squared_error(y_test,y_pred))
print('Accuracy: ',accuracy)
print('RMSE: ',rmse)
print('MAE: ',mae)
asd=classification_report(y_pred,y_test)
print(asd)
```
