```python
import pandas as  pd
import missingno as msno
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn import preprocessing
```

```python
df=pd.read_csv('../data/data.csv')
```

```python
msno.matrix(df.sample(100))
```

```python
msno.bar(df.sample(100))
```

```python
corr_df = df.corr(method='spearman')
mask = np.zeros_like(corr_df)
mask[np.triu_indices_from(mask)] = True
# {pearson, kendall, spearman}
print("--------------- CREATE A HEATMAP ---------------")
sns.set(font_scale=2)
plt.figure(figsize = (30,25))
with sns.axes_style("white"):
    sns_plot = sns.heatmap(corr_df,mask=mask,cmap='RdYlGn_r', vmax=1.0, vmin=-1.0,annot=True,annot_kws={"size": 20},square=True)
fig = sns_plot.get_figure()
fig.savefig("../images/heatmap.png")
```
